```python
import numpy as np, matplotlib.pyplot as plt, pandas as pd
from scipy.integrate import solve_ivp
```

```python
# Padrão
def sir(t, y):
    
    s, i, r = y    
    dsdt = - R0 * s * i
    didt = R0 * s * i - i
    drdt = i

    return dsdt, didt, drdt


t0, t = 0, 100
i0 = 1e-10
r0 = 0
s0 = 1 - i0 - r0
R0 = 2

res = solve_ivp(sir, (t0, t), y0=[s0, i0, r0], method='LSODA')

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(res.t, res.y[0],'g',label='Sus')
ax.plot(res.t, res.y[1],'r',label='Inf')
ax.plot(res.t, res.y[2],'b',label='Rec')
ax.legend( frameon=False)
```

```python
#MAscarados

def sir(t, y, R0, m):
    
    s, i, r, m = y   
    
    dsdt = - R0 * s * i - mascara*i + 0.01*m
    
    dmdt = mascara*i - 0.01*m
    
    didt = R0 * s * i - i
    drdt = i 
    return dsdt, didt, drdt, dmdt


t0, t = 0, 100
i0 = 1e-5
r0 = 0
m0 = 0
o0 = 0
s0 = 1 - i0 - r0 - m0

mascara = 0.01
R0 = 1.5

res = solve_ivp(sir, (t0, t), y0=[s0, i0, r0, m0],args=(R0,m), method='LSODA')

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(res.t, res.y[0],'g',label='Sus')
ax.plot(res.t, res.y[1],'r',label='Inf')
ax.plot(res.t, res.y[2],'b',label='Rec')
ax.plot(res.t, res.y[3],label='Mas')

ax.set_ylim([0,1e-2])
ax.legend( frameon=False)
```

```python
#MAscarados

def sir(t, y):
    
    s, i, r, m = y   
    
    consciencia = i*0.1
    relaxamento = 0.1
    R0 = 2
    
    dsdt = - R0 * s * i - consciencia 
    dmdt = consciencia - m*i*relaxamento
    didt = R0 * s * i + R0*m*i*relaxamento - i
    drdt = i
    
    return dsdt, didt, drdt, dmdt

t0, t = 0, 20
i0 = 1e-5
r0 = 0
m0 = 0
o0 = 0
s0 = 1 - i0 - r0 - m0

res = solve_ivp(sir, (t0, t), y0=[s0, i0, r0, m0], method='LSODA')

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(res.t, res.y[0],'g',label='Sus')
ax.plot(res.t, res.y[1],'r',label='Inf')
ax.plot(res.t, res.y[2],'b',label='Rec')
ax.plot(res.t, res.y[3],label='Mas')

ax.set_ylim([0,1e-0])
ax.legend( frameon=False)
```
