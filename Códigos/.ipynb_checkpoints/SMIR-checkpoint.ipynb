```python
import numpy as np, matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def smir(t, y, β, ε, δ):
    s, m, i, r = y
    dsdt = - β*s*i - δ*s
    dmdt = δ*s - ε*m*i
    didt = β*s*i + ε*m*i - i
    drdt = i
    return dsdt, dmdt, didt, drdt

β = 1.6
δ = 0.002
ε = δ/5

t0, t = 0, 100
i0 = 1e-10
r0 = 0
m0 = 0
s0 = 1 - m0 - i0 - r0

t_discreto = np.linspace(t0,t,t)

res = solve_ivp(smir, (t0, t), y0=[s0, m0, i0, r0], args=( β, ε, δ), t_eval=t_discreto, method='LSODA')

fig, ax = plt.subplots(figsize=(10,6))
ax.plot(res.t, res.y[0], 'g', label='Sucetíveis')
ax.plot(res.t, res.y[1], 'b', label='Mascarados')
ax.plot(res.t, res.y[2], 'r', label='Infectados')
ax.plot(res.t, res.y[3], 'gray', label='Recuperados')

ax.grid(True)
ax.set_ylabel('Densidade populacional')
ax.set_xlabel('Dias')
ax.legend(frameon='False')
ax.set_ylim([0,1e-1])

```

```python
# SMIR + tanh

def smir(t, y, β, ε):
    s, m, i, r = y
    
    δ = - A*np.tanh(t-t_δ) + M
    ε = - B*np.tanh(t-t_ε) + N
    
    dsdt = - β*s*i - δ*s
    dmdt = δ*s - ε*m*i
    didt = β*s*i + ε*m*i - i
    drdt = i
    
    return dsdt, dmdt, didt, drdt


δ0 = 0
δ  = 0.1
M = (δ0+δ)/2
A = δ0 - M
t_δ = 15

ε0 = 0
ε  = 1.2
N = (ε0+ε)/2
B = ε0 - N
t_ε = 23

β = 1.5

t0, t = 0, 40
i0 = 1e-5
r0 = 0
m0 = 0
s0 = 1 - m0 - i0 - r0

t_discreto = np.linspace(t0,t,1000)

res = solve_ivp(smir, (t0, t), y0=[s0, m0, i0, r0], args=( β, ε), t_eval=t_discreto, method='LSODA')

fig, ax = plt.subplots(figsize=(10,6))
ax.plot(res.t, res.y[0], 'g', label='Sucetíveis')
ax.plot(res.t, res.y[1], 'b', label='Mascarados')
ax.plot(res.t, res.y[2], 'r', label='Infectados')
ax.plot(res.t, res.y[3], 'gray', label='Recuperados')

ax.grid(True)
ax.set_ylabel('Densidade populacional')
ax.set_xlabel('Dias')
ax.legend(frameon='False')
ax.set_ylim([0,1e-1])

```

```python
# SWIR + tanh

def swir(t, y, β, ε, δ):
    s, e, i, r = y

    dsdt = - β*s*i - δ*s*i
    dedt = δ*s*i - ε*e
    didt = β*s*i + ε*e - i
    drdt = i
    
    return dsdt, dedt, didt, drdt

δ = 0.5
ε = 0.7
β = 1.1

t0, t = 0, 60
i0 = 1e-6
r0 = 0
e0 = 0
s0 = 1 - e0 - i0 - r0

t_discreto = np.linspace(t0,t,1000)

res = solve_ivp(swir, (t0, t), y0=[s0, e0, i0, r0], args=( β, ε, δ), t_eval=t_discreto, method='LSODA')

fig, ax = plt.subplots(figsize=(10,6))
ax.plot(res.t, res.y[0], 'g', label='Sucetíveis')
ax.plot(res.t, res.y[1], 'b', label='Expostos')
ax.plot(res.t, res.y[2], 'r', label='Infectados')
ax.plot(res.t, res.y[3], 'gray', label='Recuperados')

ax.grid(True)
ax.set_ylabel('Densidade populacional')
ax.set_xlabel('Dias')
ax.legend(frameon='False')
ax.set_ylim([0,1e-1])

```
