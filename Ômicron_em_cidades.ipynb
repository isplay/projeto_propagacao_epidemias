```python
import numpy as np, matplotlib.pyplot as plt, pandas as pd

# Labels do eixo horizontal considerando o período de início de fim da coleta de dados.
lista = []
for ano in range(2020,2023,1):
    for mes in range(1,13,1):
        if mes < 10:
            lista.append(f'{ano}/0{mes}/01')
        else:
            lista.append(f'{ano}/{mes}/01')
labels_lista = lista[3:28]
```

# Fonte dos dados: 
Dados do Estado do Rio de Janeiro:

Base de Datos COVID-19 (Chile): https://www.minciencia.gob.cl/covid19/ e https://github.com/MinCiencia/Datos-COVID19

Base de Datos COVID-19 (USA): https://usafacts.org/visualizations/coronavirus-covid-19-spread-map

Casos confirmados em Buenos Aires - Argentina: <https://data.buenosaires.gob.ar/dataset/casos-covid-19>



# Niterói - Brasil

```python
dados0 = pd.read_csv('/home/isplay/projeto_propagacao_epidemias/csv/Infectados_em_Niterói.csv', index_col=0)

população_niteroi = 516981
# Fonte: <https://www.ibge.gov.br/cidades-e-estados/rj/niteroi.html> 28/04/2022

dados0_y_niteroi = dados0['Niteroi'].to_numpy()/população_niteroi
dados0_x_niteroi = dados0.index.to_numpy()

fig, ax = plt.subplots(figsize=(15,8))
ax.plot(dados0_x_niteroi, dados0_y_niteroi, label='Casos confirmados')
ax.grid(True)
ax.set_ylabel('Densidade populacional')
ax.legend(frameon=False)
ax.set_xticks(labels_lista)
fig.autofmt_xdate(rotation=45)
plt.title(f'Casos confirmados na cidade Niterói; População de {população_niteroi} habitantes.')
```

# Buenos Aires - Argentina

```python
dados0 = pd.read_csv('/home/isplay/projeto_propagacao_epidemias/csv/buenos_aires_casos.csv', index_col=1)

população_ba = 2891082
# Fonte: https://pt.wikipedia.org/wiki/Buenos_Aires - 28/04/2022

dados0_y_ba = dados0['Casos confirmados'].to_numpy()/população_ba
dados0_x_ba = dados0.index.to_numpy()

fig, ax = plt.subplots(figsize=(15,8))
ax.plot(dados0_x_ba, dados0_y_ba, label='Casos confirmados')
ax.grid(True)
ax.set_ylabel('Densidade populpopacional')
ax.legend(frameon=False)
ax.set_xticks(labels_lista)
fig.autofmt_xdate(rotation=45)
plt.title(f'Casos confirmados na cidade Buenos Aires; População de {população_ba} habitantes.')
```
