```python
import numpy as np, matplotlib.pyplot as plt, pandas as pd
from scipy.integrate import solve_ivp, odeint
```

# Pico da Ômicron

```python
população = 515317 # 2020

# Coletando os dados.

url = 'https://raw.githubusercontent.com/isplay/projeto_propagacao_epidemias/main/csv/Infectados_em_Niter%C3%B3i.csv'
niteroi = pd.read_csv(url, index_col=0)
```

```python
# Intervalo de tempo em que a ômicron se propagou.

omi = niteroi['2021/12/01':'2022/03/01']


# Labels do eixo horizontal.

lista = []
for ano in range(2020,2023,1):
    for mes in range(1,13,1):
        if mes < 10:
            lista.append(f'{ano}/0{mes}/01')
        else:
            lista.append(f'{ano}/{mes}/01')

labels_lista = []
for i in omi.index:
    if i in lista:
        labels_lista.append(i)

# Dados dos eixos.

dias = omi.index
dias.to_numpy()
        
omi['Casos'] = omi['Casos']/população
casos = omi['Casos'].to_numpy()


# Plot.

fig, ax = plt.subplots(figsize=(15,8))
ax.plot(dias, casos,'k', label='Casos confirmados')
ax.set_xticks(labels_lista)
ax.grid(True)
ax.set_xlabel('Tempo (Meses)')
ax.set_ylabel('População (%)')
ax.legend(frameon=False, fontsize=13)
ax.set_title(f'COVID19 em Niterói entre os dias {dias[0]} e {dias[-1]}.')  
```

```python
# SIR.

# sub de subida.
# des de descida.
# Gamma = 1

def sir_sub(y, t):
    s, i, r = y
    dsdt = - R0_sub * s * i
    didt = R0_sub * s * i - i
    drdt = i
    return dsdt, didt, drdt

def sir_des(y, t):
    s, i, r = y
    dsdt = - R0_des * s * i
    didt = R0_des * s * i - i
    drdt = i
    return dsdt, didt, drdt



R0_sub = 1.123
i0_sub = 1e-5
r0_sub = 0
s0_sub = 1 - i0_sub - r0_sub

t_sub = np.linspace(0,44)
y0_sub = s0_sub, i0_sub, r0_sub
s_sub, i_sub, r_sub = odeint(sir_sub, y0_sub, t_sub).T


R0_des = 0.954
i0_des = i_sub[-1]
r0_des = r_sub[-1]
s0_des = 1 - i0_des - r0_des

t_des = np.linspace(44,86)
y0_des = s0_des, i0_des, r0_des
s_des, i_des, r_des = odeint(sir_des, y0_des, t_des).T



# Plot.

fig, ax = plt.subplots(figsize=(16,8))

ax.plot(t_sub, s_sub, '--g', label='Suscetível')
ax.plot(t_sub, i_sub, 'o-r',lw=3, label='Infectados')
ax.plot(t_sub, r_sub, '-b', label='Recuperados')

ax.plot(t_des, s_des, '--g')
ax.plot(t_des, i_des, 'o-r', lw=3)
ax.plot(t_des, r_des, '--b')

ax.plot(dias,casos,'k',label='Infec. Niterói')
plt.axvline(x = '2022/01/17', color='purple', lw=1, label = f"Pico no dia 2022/01/17")

ax.legend(frameon=False, loc='upper right', fontsize=13)
ax.set_xticks(labels_lista)
ax.set_xlabel('Tempo (Meses)')
ax.set_ylabel('População (%)')
ax.grid(True)
ax.set_ylim([0, 0.0025])
ax.text( 0.8, 0.6, f'R0 = {R0_sub} e {R0_des}\ni0 = {i0_sub}\nPopulação = {população}', transform=ax.transAxes, fontsize=13)
plt.title('Pico da Ômicron: SIR')
```

```python
# SIR: 2 R0.

def sir(y, t, t_pico, R0_sub, R0_des):
    s, i, r = y
    if t <= t_pico:
        dsdt = - R0_sub * s * i
        didt = R0_sub * s * i - i
        drdt = i
    else:
        dsdt = - R0_des * s * i
        didt = R0_des * s * i - i
        drdt = i
    return dsdt, didt, drdt


t_pico = 44
R0_sub = 1.123
R0_des = 0.954

i0 = 1e-5
r0 = 0
s0 = 1 - i0 - r0

t = np.linspace(0,86)
y0 = s0, i0, r0
s, i, r = odeint(sir, y0, t, args=(t_pico, R0_sub, R0_des)).T


# Plot.

fig, ax = plt.subplots(figsize=(16,8))

ax.plot(t, s, '--g', label='Suscetível')
ax.plot(t, i, 'o-r',lw=3, label='Infectados')
ax.plot(t, r, '-b', label='Recuperados')


ax.plot(dias,casos,'k',label='Infec. Niterói')
plt.axvline(x = '2022/01/17', color='purple', lw=1, label = f"Pico no dia 2022/01/17")
ax.legend(frameon=False, loc='upper right', fontsize=13)
ax.set_xticks(labels_lista)
ax.set_xlabel('Tempo (Meses)')
ax.set_ylabel('População (%)')
ax.grid(True)
ax.set_ylim([0, 0.0025])
ax.text( 0.8, 0.6, f'R0 = {R0_sub} e {R0_des}\ni0 = {i0_sub}\nPopulação = {população}', transform=ax.transAxes, fontsize=13)
plt.title('Pico da Ômicron: SIR')

```

```python
# SEIR.

# sub de subida.
# des de descida.
# Gamma = 1

def seir(t, y, alpha, R0, delta):
    s, e, i, r = y
    dsdt = - R0 * s * i - delta * s * i
    dedt = delta * s * i - alpha * e * i
    didt = R0 * s * i + alpha * e * i - i
    drdt = i
    return dsdt, dedt, didt, drdt


# Subida.

R0_sub = 1.135
alpha_sub = 1
delta_sub = R0_sub + 0.2
i0_sub = 5e-7

t0_sub, t_sub = 0, 44
e0_sub = 0
r0_sub = 0
s0_sub = 1 - e0_sub - i0_sub - r0_sub

res_sub = solve_ivp(seir, (t0_sub, t_sub), args=(alpha_sub, R0_sub, delta_sub), 
                y0=[s0_sub, e0_sub, i0_sub, r0_sub], method='Radau')

# Descida.

R0_des = 0
alpha_des = 10
delta_des = R0_des + 0.1

t0_des, t_des = 44, 86
e0_des = res_sub.y[1][-1]
i0_des = res_sub.y[2][-1]
r0_des = res_sub.y[3][-1]
s0_des = 1 - e0_des - i0_des - r0_des

res_des = solve_ivp(seir, (t0_des, t_des), args=(alpha_des, R0_des, delta_des), 
                y0=[s0_des, e0_des, i0_des, r0_des], method='Radau')


# Plot.

fig, ax = plt.subplots(figsize=(16,8))

ax.plot(res_sub.t, res_sub.y[0], '--g', label='Suscetível')
ax.plot(res_sub.t, res_sub.y[1], 'o-r',lw=3, label='Infectados')
ax.plot(res_sub.t, res_sub.y[2], '-b', label='Recuperados')

ax.plot(res_des.t, res_des.y[0], '--g')
ax.plot(res_des.t, res_des.y[1], 'o-r',lw=3)
ax.plot(res_des.t, res_des.y[2], '-b')

ax.plot(dias,casos,'k',label='Infec. Niterói')
plt.axvline(x = '2022/01/17', color='purple', lw=1, label = f"Pico no dia 2022/01/17")

ax.legend(frameon=False, loc='upper right', fontsize=13)
ax.set_xticks(labels_lista)
ax.set_xlabel('Tempo (Meses)')
ax.set_ylabel('População (%)')
ax.grid(True)
ax.set_ylim([0, 0.0025])
ax.text( 0.8, 0.6, f'R0 = {R0_sub} e {R0_des}\ni0 = {i0_sub}\nPopulação = {população}', transform=ax.transAxes, fontsize=13)
plt.title('Pico da Ômicron: SEIR')
```

```python
# SIS.

# sub de subida.
# des de descida.
# Gamma = 1

def sis_sub(t, y):
    s, i = y
    dsdt = - R0_sub * s * i + i
    didt = R0_sub * s * i - i
    return dsdt, didt

def sis_des(t, y):
    s, i = y
    dsdt = - R0_des * s * i + i
    didt = R0_des * s * i - i
    return dsdt, didt


# Subida.

R0_sub = 1.12
i0_sub = 1e-5

t0_sub, t_sub = 0, 44
s0_sub = 1 - i0_sub 

res_sub = solve_ivp(sis_sub, (t0_sub, t_sub), y0=[s0_sub, i0_sub], method='Radau')


# Descida.

R0_des = 0.92
i0_des = res_sub.y[1][-1]

t0_des, t_des = 44, 86
s0_des = 1 - i0_des

res_des = solve_ivp(sis_des, (t0_des, t_des), y0=[s0_des, i0_des], method='Radau')


# Plot.

fig, ax = plt.subplots(figsize=(16,8))

ax.plot(res_sub.t, res_sub.y[0], '--g', label='Suscetível')
ax.plot(res_sub.t, res_sub.y[1], 'o-r',lw=3, label='Infectados')

ax.plot(res_des.t, res_des.y[0], '--g')
ax.plot(res_des.t, res_des.y[1], 'o-r',lw=3)

ax.plot(dias,casos,'k',label='Infec. Niterói')
plt.axvline(x = '2022/01/17', color='purple', lw=1, label = f"Pico no dia 2022/01/17")

ax.legend(frameon=False, loc='upper right', fontsize=13)
ax.set_xticks(labels_lista)
ax.set_xlabel('Tempo (Meses)')
ax.set_ylabel('População (%)')
ax.grid(True)
ax.set_ylim([0, 0.0025])
#ax.text( 0.8, 0.6, f'R0 = {R0_sub} e {R0_des}\ni0 = {i0}\nPopulação = {população}', transform=ax.transAxes, fontsize=13)
plt.title('Pico da Ômicron: SIS')
```

```python
# SIRD.

# sub de subida.
# des de descida.
# Gamma = 1

def sird(t, y, R0, u):
    s, i, r, m = y
    dsdt = - R0 * s * i
    didt = R0 * s * i - i - u * i
    drdt = i
    dudt = u * i
    return dsdt, didt, drdt, dudt


# Subida.

R0_sub = 1.224
i0_sub = 1e-5
u_sub = 0.1

t0_sub, t_sub = 0, 44
r0_sub = 0
u0_sub = 0
s0_sub = 1 - i0_sub - r0_sub - u0_sub

res_sub = solve_ivp(sird, (t0_sub, t_sub), y0=[s0_sub, i0_sub, r0_sub, u0_sub], args=(R0_sub, u_sub),method='Radau')


# Descida.

R0_des = 1.2
u_des = 0.24

t0_des, t_des = 44, 86
i0_des = res_sub.y[1][-1]
r0_des = res_sub.y[2][-1]
u0_des = res_sub.y[3][-1]
s0_des = 1 - i0_des - r0_des - u0_des

res_des = solve_ivp(sird, (t0_des, t_des), y0=[s0_des, i0_des, r0_des, u0_des], 
                    args=(R0_des, u_des), method='Radau')


# Plot.

fig, ax = plt.subplots(figsize=(16,8))

ax.plot(res_sub.t, res_sub.y[0], '--g', label='Suscetível')
ax.plot(res_sub.t, res_sub.y[1], 'o-r', lw=3, label='Infectados')
ax.plot(res_sub.t, res_sub.y[2], '-b', label='Recuperados')
ax.plot(res_sub.t, res_sub.y[3], '-g', label='Mortes')

ax.plot(res_des.t, res_des.y[0], '--g')
ax.plot(res_des.t, res_des.y[1], 'o-r',lw=3)
ax.plot(res_des.t, res_des.y[2], '-b')
ax.plot(res_des.t, res_des.y[3], '-g')

ax.plot(dias,casos,'k',label='Infec. Niterói')
plt.axvline(x = '2022/01/17', color='purple', lw=1, label = f"Pico no dia 2022/01/17")

ax.legend(frameon=False, loc='upper right', fontsize=13)
ax.set_xticks(labels_lista)
ax.set_xlabel('Tempo (Meses)')
ax.set_ylabel('População (%)')
ax.grid(True)
ax.set_ylim([0, 0.0025])
#ax.text( 0.8, 0.6, f'R0 = {R0_sub} e {R0_des}\ni0 = {i0}\nPopulação = {população}', transform=ax.transAxes, fontsize=13)
plt.title('Pico da Ômicron: SIRD')
```

```python
# SIRV.

# sub de subida.
# des de descida.
# Gamma = 1

def sirv(t, y, R0, v):
    s, i, r, m = y
    dsdt = - R0 * s * i - v * s
    didt = R0 * s * i - i 
    drdt = i
    dvdt = v * s
    return dsdt, didt, drdt, dvdt


# Subida.

R0_sub = 1.227
i0_sub = 1e-6
v_sub = 0.002

t0_sub, t_sub = 0, 44
r0_sub = 0
v0_sub = 0
s0_sub = 1 - i0_sub - r0_sub - v0_sub

res_sub = solve_ivp(sirv, (t0_sub, t_sub), y0=[s0_sub, i0_sub, r0_sub, v0_sub], args=(R0_sub, v_sub),method='Radau')


# Descida.

R0_des = 1.05
v_des = 0.001

t0_des, t_des = 44, 86
i0_des = res_sub.y[1][-1]
r0_des = res_sub.y[2][-1]
v0_des = res_sub.y[3][-1]
s0_des = 1 - i0_des - r0_des - v0_des

res_des = solve_ivp(sirv, (t0_des, t_des), y0=[s0_des, i0_des, r0_des, v0_des], 
                    args=(R0_des, v_des), method='Radau')


# Plot.

fig, ax = plt.subplots(figsize=(16,8))

ax.plot(res_sub.t, res_sub.y[0], '--g', label='Suscetível')
ax.plot(res_sub.t, res_sub.y[1], 'o-r', lw=3, label='Infectados')
ax.plot(res_sub.t, res_sub.y[2], '-b', label='Recuperados')
ax.plot(res_sub.t, res_sub.y[3], '-y', label='Vacinados')

ax.plot(res_des.t, res_des.y[0], '--g')
ax.plot(res_des.t, res_des.y[1], 'o-r',lw=3)
ax.plot(res_des.t, res_des.y[2], '-b')
ax.plot(res_des.t, res_des.y[3], '-y')

ax.plot(dias,casos,'k',label='Infec. Niterói')
plt.axvline(x = '2022/01/17', color='purple', lw=1, label = f"Pico no dia 2022/01/17")

ax.legend(frameon=False, loc='upper right', fontsize=13)
ax.set_xticks(labels_lista)
ax.set_xlabel('Tempo (Meses)')
ax.set_ylabel('População (%)')
ax.grid(True)
ax.set_ylim([0, 0.0025])
ax.text( 0.8, 0.6, f'R0 = {R0_sub} e {R0_des}\ni0 = {i0_sub}\nPopulação = {população}', transform=ax.transAxes, fontsize=13)
plt.title('Pico da Ômicron: SIRV')
```
