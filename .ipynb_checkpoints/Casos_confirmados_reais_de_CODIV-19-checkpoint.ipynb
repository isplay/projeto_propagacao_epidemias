```python
lista = arr
for i in range(len(arr)):
    b = lista[i].split('/')
    b[0], b[-1] = b[-1], b[0]
    b[1], b[-1] = b[-1], b[1]
    c = '/'.join(b)
    lista[i] = c
lista
```

```python
lista0 = lista
for i in range(len(lista0)):
    b = lista0[i].split('/')
    if b[0] == '20':
        b[0] = '2020'
    if b[0] == '21':
        b[0] = '2021'
    if b[0] == '22':
        b[0] = '2022'
    c = '/'.join(b)
    lista0[i] = c
lista0
```

```python
lista1 = lista0
for i in range(len(lista1)):
    b = lista1[i].split('/')
    if b[1] in ['1','2','3','4','5','6','7','8','9']  and '0' not in b[1]:
        b[1] = '0' + b[1]
    c = '/'.join(b)
    lista1[i] = c
lista1
```
